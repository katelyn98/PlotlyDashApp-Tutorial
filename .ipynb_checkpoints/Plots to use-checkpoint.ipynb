```python
import plotly.express as px
import pandas as pd
```

```python
# data sourced from: https://simplemaps.com/data/us-cities
```

```python
df = pd.read_csv('simplemaps_data/uscities.csv')
df = df.head(100)

fig = px.scatter_geo(df, lat="lat", lon="lng", color="timezone",
                    hover_name="city",
                    hover_data={
                        "lng":False,
                        "timezone": False,
                        "lat":False,
                        "state_name": True,
                        "state_id": True,
                        "population": True
                    })
fig.update_layout(title = '100 US Cities', geo_scope="usa")
fig.show()
```

```python
df = px.data.gapminder()
fig = px.area(df, x="year", y="pop", color="continent",line_group="country")

fig.update_layout(
    xaxis=dict(
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)
fig.show()
```
